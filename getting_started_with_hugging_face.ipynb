# Pipeline

```python
from transformers import pipeline
```

**Pipelines documentation with Available Pipelines**

* https://huggingface.co/docs/transformers/main_classes/pipelines#pipelines

```python
# sentiment analysis

classifier = pipeline("sentiment-analysis")

res = classifier("EFSA is a great place to work")

print(res)
```

```python
# text-generation

generator = pipeline("text-generation", model="distilgpt2")

res = generator(
    "In this course, we will teach you how to",
    max_length = 30,
    num_return_sequences=2
)

print(res)
```

```python
# zero-shot-classification

classifier = pipeline("zero-shot-classification")

res = classifier(
    "This is a course about Python list comprehension",
    candidate_labels=["education", "politics", "business"]
)

print(res)
```

# Tokenizer & Model
* Tokenization breaks text into smaller parts for easier machine analysis, helping machines understand human language

```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Getting the pipeline without the tokenizer

classifier = pipeline("sentiment-analysis")

res = classifier("EFSA is a great place to work")

print(res)

# tokenizer

model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Applying the pipeline with tokenizer

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

res = classifier("EFSA is a great place to work")

print(res)
```

```python
# Taking a look at the tokenizer and see what it's doing
sequence = "Using a Transfomer network is simple"
res = tokenizer(sequence)
print(res)
tokens = tokenizer.tokenize(sequence)
print(tokens)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
decoded_string = tokenizer.decode(ids)
print(decoded_string)
```

# Applying PyTorch or Tensorflow

```python
from transformers import pipeline
from transformers import AutoTokenizer, TFLayoutLMMainLayer
import torch
import torch.nn.functional as F
```

```python
# Getting the tokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Applying the pipeline with tokenizer just to compare
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Getting the training data (note: this is an example usually this is applied with more data)
X_train = ["EFSA is a great place to work"]

res = classifier(X_train)

print(res)

# Applying Pytorch
batch = tokenizer(X_train, padding=True, truncation=True, max_length=512, return_tensors="pt")
print(batch)

with torch.no_grad():
  outputs = model(**batch)
  print(outputs)
  predictions = F.softmax(outputs.logits, dim=1)
  print(predictions)
  labels = torch.argmax(predictions, dim=1)
  print(labels)
```
